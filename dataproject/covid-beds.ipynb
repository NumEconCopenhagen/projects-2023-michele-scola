```python
cov = pd.read_csv('data.csv')
cov.head(15)
```

```python
drop_these = ['Lat', 'Long','Province/State'] # use list comprehension to create list of columns
cov.drop(drop_these, axis=1, inplace=True) # axis = 1 -> columns, inplace=True -> changed, no copy made
cov.head(15)
```

```python
cov2 = cov.groupby(cov['Country/Region']).aggregate(np.sum)
cov2.head(15)
```

```python
#cov2.convert_dtypes()
cov2.rename(index = {'Country/Region':'Country'}, inplace=True)
cov2.reset_index()
```

```python
cov2.info()
```

```python
print(cov2.columns)
```

```python
cov_long = pd.wide_to_long(df=cov2, stubnames='death',i='Country/Region', j='day')
cov_long.reset_index(inplace=True)
cov_long.head(5)
```

```python
import ipywidgets as widgets
def plot(cov, country): 
    I = cov['Country/Region'] == country
    ax=cov.loc[I,:].plot(x='day', y='death', style='-o', legend=False)
```

```python
widgets.interact(plot, 
    df = widgets.fixed(cov_long),
    country = widgets.Dropdown(description='country', 
                                    options=cov_long.'Country/Region'.unique(), 
                                    value='Afghanistan')
); 
```

# Dataset 2

```python
wb_beds = wb.download(indicator='SH.MED.BEDS.ZS', start=1960, end=2017)
wb_beds = wb_beds.rename(columns = {'SH.MED.BEDS.ZS':'beds'})
wb_beds = wb_beds.reset_index()
wb_beds.head(10)
```

```python
wb_beds.year = wb_beds.year.astype(int) # convert year
wb_beds.country = wb_beds.country.astype('string') # convert country to the special pandas string type
wb_beds.info()
```

```python
wb_beds2 = wb_beds.pivot_table('beds','country','year')
wb_beds2
```

```python
import ipywidgets as widgets
def plot(bed, country): 
    I = bed['country'] == country
    ax=bed.loc[I,:].plot(x='year', y='beds', style='-o', legend=False)
```

```python
widgets.interact(plot, 
    bed = widgets.fixed(wb_beds),
    country = widgets.Dropdown(description='country', 
                                    options=wb_beds.country.unique(), 
                                    value='Canada')
); 
```
